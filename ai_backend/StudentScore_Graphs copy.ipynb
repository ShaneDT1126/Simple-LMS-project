```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import mysql.connector

data = pd.read_csv("C:/SchoolProject/Simple-LMS-project/ai_backend/data/student-scores.csv", encoding="ISO-8859-1")

print(data.head())

print(data.describe())

print(data.isnull().sum())

sns.pairplot(data)
plt.show()


print("\nPlotting pie chart for the number of IT and CS students...")
department_counts = data['department'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.0f', startangle=140)
plt.title('Distribution of Students in IT and CS')
plt.show()

print("\nPlotting pie chart for the quiz scores of students...")
quiz_score = data['quiz_score'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(quiz_score, labels=quiz_score.index, autopct='%1.0f', startangle=140)
plt.title('Distribution of Quiz Score of Students')
plt.show()

print("\nPlotting pie chart for the lesson scores of students...")
lesson_score = data['lesson_score'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(lesson_score, labels=lesson_score.index, autopct='%1.0f', startangle=140)
plt.title('Distribution of Lesson Score of Students')
plt.show()
```
